```python
# importing packages
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

############# Main Section ############
# loading dataset using seaborn
df = pd.read_csv('heart.csv')

# List of independent variables (excluding the 'target' column)
independent_vars = [col for col in df.columns if col != 'HeartDisease']

# Create a figure and axis to plot
plt.figure(figsize=(20, 5))

# Loop through each independent variable and create a boxplot
for i, feature in enumerate(independent_vars):
    plt.subplot(2, 6, i+1)  # Adjust the subplot size (3x3 in this case for 9 features)
    if df[feature].dtype == 'object' or pd.api.types.is_categorical_dtype(df[feature]):
        sns.barplot(x=feature, y='HeartDisease', data=df)
        plt.ylabel("Heart Disease Proportion")
    else:
        sns.kdeplot(df, x=feature, hue="HeartDisease", fill=True)
    #sns.boxplot(x='HeartDisease', y=feature, data=df)
    plt.title(f'Heart disease by {feature}')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
